```python
# Funcion para explorar las caracteristicas de los DataFrames
def describe_exploration(df):
    
    s=df.shape
    t=df.dtypes
    d=df.describe()
    d_o=df.describe(include='object')
    
    return s, t, d, d_o

#Exploracion de la informacion de las columnas
def columns_exploration(df,column_name):
    print(df[column_name].value_counts())
    

# Funcion para explorar de los valores nulos
def null_exploration(df):
    
    print(df.isnull().sum())
    print(df.isnull().mean())
    
# Funcion para arrojar las columnas que considero irrelevantes parta 
def drop_bas(df, list_obj):
    drop_var=[]
    for i in df.columns:
        if i not in list_obj:
            drop_var.append(i)
    df = df.drop(drop_var, axis=1)
    return df
    
# Cambio de los valores nulos
def null_imputation(df,column_name):
    df[column_name].fillna(df[column_name].median(), inplace=True)
    
    return df

def hist_graph(df, column_names):
    for column_name in column_names:

        df[column_name].hist(bins=30, figsize=(30,15), density=True, histtype='step',linewidth=2)
        plt.xticks(rotation='vertical')

        plt.title(column_name)
        plt.show()


def hori_boxplot(y):
    sns.set_theme(style="ticks")

    # Initialize the figure with a logarithmic x axis
    f, ax = plt.subplots(figsize=(20, 17))
    ax.set_xscale("log")

    # dataset
    sb

    # Plot the orbital period with horizontal boxes
    sns.boxplot(x="PUNT_GLOBAL", y=y, data=sb,
                whis=[0, 100], width=.4, palette="vlag")

    # Add in points to show each observation
    sns.stripplot(x="PUNT_GLOBAL", y=y, data=sb,
                  size=3, color=".3", linewidth=0)

    # Tweak the visual presentation
    ax.xaxis.grid(True)
    ax.set(ylabel="")
    sns.despine(trim=True, left=True)
    return plt.show()

def change_date_format(df, new_column_name, column_name):
    df[new_column_name] = pd.to_datetime(df[column_name], format='%Y%m')
    df["{}_Year".format(new_column_name)] = df[new_column_name].dt.year
    df["{}_Month".format(new_column_name)] = df[new_column_name].dt.month
    
    return df


def groupby_2_cols_against_target(df, column1, column2, target):

    groupby_2_cols_against_target = df.groupby(
        [column1, column2]
    )[target].sum() / df.groupby(
        [column1, column2]
    )[target].count() * 100.0

    pivot_t = groupby_2_cols_against_target.reset_index().pivot(
        index=column1,
        columns=column2,
        values=target
    )
    
    return pivot_t


def groupby_var_target(df, column_name, target_name, title):
    ## group by
    groupby_var_target = df.groupby(
        column_name
    )[target_name].sum() / df.groupby(
        column_name
    )[target_name].count() * 100.0
    
    ## plot groupby
    ax = groupby_var_target.sort_values(ascending=False).plot(
        kind='bar',
        color='skyblue',
        grid=True,
        figsize=(10, 7),
        title=title
    )

def groupby_var_percent(df, column_name, target_name, title):
    ## group by
    groupby_var_target = df.groupby(
        column_name
    )[target_name].sum() / df.groupby(
        column_name
    )[target_name].count() 
    
    ## plot groupby
    ax = groupby_var_target.sort_values(ascending=False).plot(
        kind='bar',
        color='skyblue',
        grid=True,
        figsize=(10, 7),
        title=title
    )
    
    return groupby_var_target, plt.show()

## to plot hist to all dataframe
def corr_matrix(df, column_names):
    corr = df[column_names].corr()
    plt.figure(figsize=(23,13))

    sns.heatmap(corr,
                xticklabels=corr.columns.values,
                yticklabels=corr.columns.values,annot=True)

    corr
    
def scatter_plot_target(df,column_names, target):
    for column_name in column_names:
        plt.scatter(df[column_name], df[target])
        plt.title(column_name)
        plt.show()
        
def histograma_esparcido(x,y,files,titulos):


    
    fig, ax = plt.subplots(figsize=(10.5,7))

    ax.scatter(x, y, c='g', alpha=0.5)
    
    for i in range(len(files)):
        if files[i]==x:
            plt.xlabel(f'{titulos[i]}')
        if files[i]==y:
            plt.ylabel(f'{titulos[i]}')


    ax.set_aspect('auto')

    divider = make_axes_locatable(ax)

    ax_histx = divider.append_axes("top", 2.1, pad=0.1, sharex=ax)
    ax_histy = divider.append_axes("right", 2.1, pad=0.1, sharey=ax)

    ax_histx.xaxis.set_tick_params(labelbottom=False)
    ax_histy.yaxis.set_tick_params(labelleft=False)

    binwidth = 0.25
    xymax = len(estrato), len(estrato)


    bins = 5
    ax_histx.hist(x, bins=bins)
    ax_histy.hist(y, bins=bins, orientation='horizontal')
    
    ax_histx.set_yticks([0, len(x)//2, len(x)])
    ax_histy.set_xticks([0, len(y)//2, len(y)])


    return plt.show(), print(f'Este panel muesta la relaciÃ³n la muestra x con la muestra y')



def dummy_encoding(df, column_name):
    column_encoded = pd.get_dummies(df[column_name])
    column_encoded.columns = ['{}_{}'.format(column_name ,x) for x in column_encoded.columns]

    df = pd.concat([df, column_encoded], axis=1)
    return df


    
    ## function to find outliers limits
def find_limits_and_delete_outliers(df, variable):

        IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

        lower_limit = df[variable].quantile(0.25) - (IQR * 1.5)
        upper_limit = df[variable].quantile(0.75) + (IQR * 1.5)

        ## to find and delete values upper and under the outliers limits
        np.where(df[variable] > upper_limit)
        np.where(df[variable] < lower_limit)
        
        outliers = np.where(df[variable] > upper_limit, True, 
                            np.where(df[variable] < lower_limit, True, False))
        
        outliers_df = df.loc[outliers, variable]
        outliers_df_ids = df.loc[outliers]
        new_df = df[variable].drop(outliers_df_ids.index).reset_index(drop=True)
        
        outliers_percentage = len(outliers_df) / len(df[variable])
        
        return outliers_percentage, new_df


def outliers_graph(df, column_names):
    for column_name in column_names:
        df[column_name].plot.box()
        plt.show()
```
